```python
import os
os.chdir('..')

import warnings
warnings.filterwarnings('ignore')
```

```python
from dotenv import load_dotenv
load_dotenv()

import mlflow
import pandas as pd
```

```python
mlflow.set_tracking_uri(os.environ['MLFLOW_TRACING_URI'])
mlflow.set_experiment("dialogue_summarization_v2")

experiment_id = mlflow.get_experiment_by_name("dialogue_summarization_v2").experiment_id
```

```python
from mlflow.tracking import MlflowClient
client = MlflowClient()
```

```python
file_path = "datasets/dialog_summarization/CSV/test.csv"
dialogue_summarization_df = pd.read_csv(file_path)
dialogue_summarization_df
```

```python
input_cols = ["id", "dialogue"]
expectation_cols = ["summary", "topic"]

def func(row):
    return {
        "inputs": row[input_cols].to_dict(),
        "expectations": row[expectation_cols].to_dict(),
    }

records = dialogue_summarization_df.apply(func, axis=1)
records
```

```python

dialogue_summarization_dataset = client.create_dataset(
    name="dialogue_summarization",
    experiment_id=experiment_id,
)
dialogue_summarization_dataset.merge_records(records.tolist()[:])
```
